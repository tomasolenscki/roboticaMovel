# transformação IMU (x´,y´,z´) → referencial local do robô (x,y,z)



```python

import numpy as np

# Parâmetros de calibração no notebook calibracao_imu.ipynb (res2.x)
SCALE = 0.000598
VIES = np.array([-109.79370126, -104.46455324, -253.01762059])
GANHOS = np.array([0.99708971, 0.99872567, 0.99779582])
```

```python

path = "/home/mateus/poli_26/elem_robotica/robotica_movel/aula5"
g1 = np.load(path+"/dados_imu_teste_normalizados_1.npy")
g2 = np.load(path+"/dados_imu_teste_normalizados_2.npy")
g3 = np.load(path+"/dados_imu_teste_normalizados_3.npy")
g4 = np.load(path+"/dados_imu_teste_normalizados_4.npy")

g1 = g1.mean(axis=0)
g2 = g2.mean(axis=0)
g3 = g3.mean(axis=0)    
g4 = g4.mean(axis=0)

g = np.mean([g1, g2, g3, g4], axis=0)
```

```python
print(g1) #gravity measured upwards
```

### eixo z em função de x´, y´, z´

```python
print(g)
k = -g/np.linalg.norm(g)    # z axis of the robot (points downwards according to convention)
print(k)
```

### eixo y em função de x´, y´, z´

```python


g5 = np.load(path+"/dados_imu_teste_normalizados_inclinado.npy")
g5 = g5.mean(axis=0)
d = g4 - g5
```

```python
# j = produto vetorial g5 x (d = g4 - g5)
j = np.cross(d, k)
j = j / np.linalg.norm(j)
print(j)
```

### eixo x em função de x´ ,y´ ,z´

```python
i = np.cross(j, k)
i = i / np.linalg.norm(i)
print(i)
```

### matrizes R_{imu}^{robô}

```python
#IMU orientada como robô - x frente, y direita e z baixo
Q_FDB = np.array([i.T, 
                  j.T, 
                  k.T])
print("Q (Frente-Direita-Baixo):")
print(Q_FDB)    
np.save(path+"/Q_FDB.npy", Q_FDB)
```

```python
Q_FEC = np.array([i.T,
                  -j.T,
                  -k.T])
print("Q (Frente-Esquerda-Cima):")
print(Q_FEC)
np.save(path+"/Q_FEC.npy", Q_FEC)
```
